```python
from ultralytics import YOLO
import matplotlib.pyplot as plt
import cv2

# Load model (có thể dùng yolov8n.pt, yolov8s.pt,... tuỳ độ lớn nhẹ)
model = YOLO('yolov8s.pt')  # bạn có thể thay bằng yolov8s.pt hoặc model bạn đã train


# Train model
model.train(
    data='/content/drive/MyDrive/Dataset/FIFA_YOLO/data.yaml',  # đường dẫn đến file YAML
    epochs=12,
    imgsz=640,
    batch=16,
    project='/content/drive/MyDrive/fifa_yolo_train',
    name='yolov8n_results',
    pretrained=True  # dùng pretrained COCO
)
```

```python
model_path = '/content/drive/MyDrive/fifa_yolo_train/yolov8n_results/weights/best.pt'

# Load model từ file trọng số
model = YOLO(model_path)
```

```python
results = model.predict('/content/drive/MyDrive/Dataset/FIFA_YOLO/test/images/VAR_17_Dutch_league_2020-mp4_frame_9000_jpg.rf.a5dac5da8103cb2d5744c0073393e26f.jpg', imgsz=640)
results[0].show() # hiển thị ảnh có kết quả nhận diện
```
