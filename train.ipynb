```python
from ultralytics import YOLO
import matplotlib.pyplot as plt
import cv2

# Load model (cÃ³ thá»ƒ dÃ¹ng yolov8n.pt, yolov8s.pt,... tuá»³ Ä‘á»™ lá»›n nháº¹)
model = YOLO('yolov8s.pt')  # báº¡n cÃ³ thá»ƒ thay báº±ng yolov8s.pt hoáº·c model báº¡n Ä‘Ã£ train


# Train model
model.train(
    data='/content/drive/MyDrive/Dataset/FIFA_YOLO/data.yaml',  # Ä‘Æ°á»ng dáº«n Ä‘áº¿n file YAML
    epochs=12,
    imgsz=640,
    batch=16,
    project='/content/drive/MyDrive/fifa_yolo_train',
    name='yolov8n_results',
    pretrained=True  # dÃ¹ng pretrained COCO
)
```

```python
model_path = '/content/drive/MyDrive/fifa_yolo_train/yolov8n_results/weights/best.pt'

# Load model tá»« file trá»ng sá»‘
model = YOLO(model_path)
```

```python
results = model.predict('/content/drive/MyDrive/Dataset/FIFA_YOLO/test/images/VAR_17_Dutch_league_2020-mp4_frame_9000_jpg.rf.a5dac5da8103cb2d5744c0073393e26f.jpg', imgsz=640)
results[0].show() # hiá»ƒn thá»‹ áº£nh cÃ³ káº¿t quáº£ nháº­n diá»‡n
```
